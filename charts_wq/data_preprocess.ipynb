```python
import pandas as pd
import numpy as np
import json
```

```python
df=pd.read_csv('yelp_all.csv')
```

```python
df
```

```python
df2=df.loc[:,['state','new_category']]
```

```python
df2['c']=1
```

```python
df2
```

```python
table = pd.pivot_table(df2, values=['c'],columns=['new_category'],
                         index =['state'],aggfunc=np.sum)
```

```python
table
```

```python
den_graph={}
```

```python
cat=table.columns
category=[]
for c in cat:
    category.append(c[1])
```

```python
total=[]
for i in range(len(category)):
    col=table.iloc[:,i]
    topV=col.sort_values(ascending=False).head(5)
    temp=[]
    for j in range(5):
        temp.append({"name":topV.index[j],"value":topV.values[j]})
    total.append({"name":category[i],"children":temp})
```

```python
den_graph["name"]="yelp"
den_graph["children"]=total
```

```python
den_graph
```

```python
type(den_graph)
```

```python
with open("yelp_dendrogram.json", "w") as outfile:
    j=json.dumps(den_graph, indent=4)
    outfile.write(j)
```

```python
outfile.close()
```

```python
bar_graph={}
temp=df2.groupby('new_category').count().state
all_level=[]
for i in range(len(temp)):
    all_level.append({'category':temp.index[i],'count':int(temp.values[i])})
```

```python
bar_graph['all']=all_level
```

```python
new_table=table.T
for i in range(27):
    temp_dict=[]
    for j in range(len(category)):
        t=np.isnan(new_table.iloc[:,i])
        new_table.iloc[:,i][t]=int(0)

        temp_dict.append({'category':category[j],'count':new_table.iloc[:,i].values[j]})

    bar_graph[new_table.columns[i]]=temp_dict
```

```python
with open("yelp_bar.json", "w") as outfile:
    j=json.dumps(bar_graph, indent=4)
    outfile.write(j)
outfile.close()
```
